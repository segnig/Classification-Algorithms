# Basics of probability

*   Probabilty defined as the chance of occurrence of an event under an experiment.





```python
import numpy as np

import scipy.stats as stats
from scipy.stats import norm, beta
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings('ignore')
```

```python
def probabilityOfHead(numberOfTrials):
    plt.figure(figsize=(8, 3))

    data = stats.bernoulli.rvs(p=0.5, size=numberOfTrials)

    x = np.linspace(0, 1, 100)

    heads = data[:numberOfTrials].sum()

    a = 1
    b = 1
    s = heads
    f = numberOfTrials - s

    y = beta.pdf(x, a + s, b + f)

    plt.plot(x, y, label=f'observe {numberOfTrials} tosses \n heads {heads}')

    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)

    plt.vlines(0.5, 0, 1, color="k", linestyles="--", lw=1)

    leg = plt.legend()

    leg.get_frame().set_alpha(0.4)

    plt.autoscale(tight=True)

    plt.title(f"Probability of getting Heads, P(head)= {heads/numberOfTrials}")

    plt.show()
```

```python
probabilityOfHead(4)
```

```python
for i in [13, 25, 56, 100, 500, 1000]:
    probabilityOfHead(i)
```
